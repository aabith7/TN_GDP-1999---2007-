```python
df_melted = dgp_tn.melt(id_vars="Year", var_name="District", value_name="GDP")
df_melted
df_gdp = df_melted.pivot(index="District", columns="Year", values="GDP")
df_gdp
df_gdp = df_gdp.drop(df_gdp.index[3])
df_gdp
```

```python
df_gdp_percentage = df.iloc[8:]
df_gdp_percentage.drop(columns = ['Description'])
df_gdp_percentage = df_gdp_percentage.melt(id_vars= 'Year' , var_name = 'District' , value_name = 'GDP')
df_gdp_percentage = df_gdp_percentage.pivot(index = 'District', columns = 'Year' , values = "GDP")
df_gdp_percentage.drop(index = "Description")
df_gdp_percentage = df_gdp_percentage.drop(df_gdp_percentage.index[3])
df_gdp_percentage
```

```python
df_gdp['total'] = df_gdp.sum(axis=1)
df_gdp['total'].sort_values(ascending=False)[0:6]
```
